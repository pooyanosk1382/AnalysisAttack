Now it's time to decrypt the cipher text.
For this we have to compare the number of letters in books and cipher text and by this way change word.

By comparing letters we can understand that L in cipher text is equals with e in real letters.
But if we want to be sure about it, we can put e in place of L in two and three letters and can see that in two letters, TL has the same plenty with he and in three letters ETL has the same plenty with the and, now we are sure about it that L is e so, we can exchange L with e.
Now we can get it that T is h and E is t.
In next cell we will change them.

```python
reader = open('cipherText.txt', 'r')
cipherText = reader.read()
cipherText = cipherText.replace('L', 'e')
cipherText = cipherText.replace('E', 't')
cipherText = cipherText.replace('T', 'h')
print(cipherText)
```
